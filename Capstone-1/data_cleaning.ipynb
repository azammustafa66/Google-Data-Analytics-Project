# This is a visualisation-cum-analysis notebook that reads CSV files, cleans the data and plots a graph of the data.

```python
import numpy as np
import pandas as pd
```

```python
# Import data
jan2023 = pd.read_csv('./2023-2024_cyclist_data/202301-divvy-tripdata.csv')
feb2023 = pd.read_csv('./2023-2024_cyclist_data/202302-divvy-tripdata.csv')
mar2023 = pd.read_csv('./2023-2024_cyclist_data/202303-divvy-tripdata.csv')
apr2023 = pd.read_csv('./2023-2024_cyclist_data/202304-divvy-tripdata.csv')
may2023 = pd.read_csv('./2023-2024_cyclist_data/202305-divvy-tripdata.csv')
jun2023 = pd.read_csv('./2023-2024_cyclist_data/202306-divvy-tripdata.csv')
jul2023 = pd.read_csv('./2023-2024_cyclist_data/202307-divvy-tripdata.csv')
aug2023 = pd.read_csv('./2023-2024_cyclist_data/202308-divvy-tripdata.csv')
sep2023 = pd.read_csv('./2023-2024_cyclist_data/202309-divvy-tripdata.csv')
oct2023 = pd.read_csv('./2023-2024_cyclist_data/202310-divvy-tripdata.csv')
nov2023 = pd.read_csv('./2023-2024_cyclist_data/202311-divvy-tripdata.csv')
dec2023 = pd.read_csv('./2023-2024_cyclist_data/202312-divvy-tripdata.csv')
jan2024 = pd.read_csv('./2023-2024_cyclist_data/202401-divvy-tripdata.csv')
feb2024 = pd.read_csv('./2023-2024_cyclist_data/202402-divvy-tripdata.csv')
mar2024 = pd.read_csv('./2023-2024_cyclist_data/202403-divvy-tripdata.csv')
apr2024 = pd.read_csv('./2023-2024_cyclist_data/202404-divvy-tripdata.csv')
may2024 = pd.read_csv('./2023-2024_cyclist_data/202405-divvy-tripdata.csv')
jun2024 = pd.read_csv('./2023-2024_cyclist_data/202406-divvy-tripdata.csv')
jul2024 = pd.read_csv('./2023-2024_cyclist_data/202407-divvy-tripdata.csv')
aug2024 = pd.read_csv('./2023-2024_cyclist_data/202408-divvy-tripdata.csv')
```

```python
df = pd.concat([jan2023, feb2023, mar2023, apr2023, may2023, jun2023, jul2023, aug2023, sep2023, oct2023, nov2023, dec2023, jan2024, feb2024, mar2024, apr2024, may2024, jun2024, jul2024, aug2024])
df.head()
```

```python
# Drop all rows with missing information
df = df.dropna()
df.head()
```

```python
# Add a column for trip duration in minutes
df['trip_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
df.head()
```

```python
# Add a column for the day of the week
df['day_of_week'] = df['started_at'].dt.day_name()
df.head()
```

```python
# Add a column for the hour of the day
df['hour_of_day'] = df['started_at'].dt.hour
df['month_of_year'] = df['started_at'].dt.month
df.head()
```

```python
# Calculate total distance by start and end latitude and longitude
df["lat_diff"] = df["end_lat"] - df["start_lat"]
df["lon_diff"] = df["end_lng"] - df["start_lng"]
df["distance"] = np.sqrt(
    (df["lat_diff"] * 111) ** 2
    + (df["lon_diff"] * 111 * np.cos(np.radians(df["start_lat"]))) ** 2
)
df.head()
```

```python
df = df.drop(
    columns=[
        "start_station_name",
        "end_station_name",
        "start_station_id",
        "end_station_id",
        "start_lat",
        "start_lng",
        "end_lat",
        "end_lng",
        "lat_diff",
        "lon_diff",
    ]
)
df.head()
```

```python
# Save the cleaned data to a new CSV file
df.to_csv("2023-2024_cyclist_data_cleaned.csv", index=False)
```
